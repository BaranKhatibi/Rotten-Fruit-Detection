```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
import math
import os
import pandas as pd
```

```python
minh = float('inf')
minw = float('inf')
for entry in os.scandir('data/all'):  
    a = entry.path
    img = cv2.imread(a)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    height, width = img_rgb.shape[:2]
    if height <= minh and width <= minw:
        minh = height
        minw = width
#         minim = img
print(minh, minw)
# plt.imshow(minim)
# plt.title('original')
# plt.show()
# h,w = minim.shape[:2]
# print(h,w)
```
