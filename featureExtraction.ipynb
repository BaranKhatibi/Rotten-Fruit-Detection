```python
import cv2
import numpy as np 
from matplotlib import pyplot as plt
import math
import os
import pandas as pd
```

```python
#finding min
minh = float('inf')
minw = float('inf')
for entry in os.scandir('data/all'):  
    a = entry.path
    img = cv2.imread(a)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    height, width = img_rgb.shape[:2]
    if height <= minh and width <= minw:
        minh = height
        minw = width

print(minh,minw)        
array_size = int(minh*minw)
print(array_size)
```

```python
names = [ 'pixel' + str(i) for i in range(array_size)]
columns = ['mean_R', 'mean_G', 'mean_B'] + names + ['label']
```

```python
avrg = ((),)
avg =  np.zeros((array_size,))
data = list()
for entry in os.scandir('data/fresh'):  
    a = entry.path
    img = cv2.imread(a)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

#     print(len(img_rgb.flatten()))
    
    dim=(minh, minw)
    resized = cv2.resize(img_rgb, dim )
    
#     print(len(resized.flatten()))
    
    rr, gg, bb = cv2.split(resized)    
    
#     print(len(rr.flatten()))
    #feature one
    r = rr.flatten()
    g = gg.flatten()
    b = bb.flatten()
    
    rm = int(r.mean())
    gm = int(g.mean())
    bm = int(b.mean())
    
#     print(len(mean))
#     print(len(r))
    #feature two

    for i in range(array_size):
        sum = r[i] + g[i] + b[i] 
        avrg = avrg +(int(sum/3),)
        print(1)
       
        
        
    data.append((rm,gm,bm,avrg,1))
    
```

```python
cnt2 = 0
for entry in os.scandir('data/rotten'):  
    a = entry.path
    img = cv2.imread(a)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    dim=(minh, minw)
    resized = cv2.resize(img_rgb, dim)
     
    
    rr, gg, bb = cv2.split(resized)
    
    r = rr.flatten()
    g = gg.flatten()
    b = bb.flatten()
    
    rm = int(r.mean())
    gm = int(g.mean())
    bm = int(b.mean())
    
    #feature two
    for i in range(array_size):
        sum = r[i] + g[i] + b[i]  
        avrg = avrg +(int(sum/3),)
        print(1)
        
        
    data.append((rm,gm,bm,avrg,2))
     
```

```python
df = pd.DataFrame(data, columns=columns)
print(df)
```

```python
import sklearn
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn import svm
from sklearn import metrics
```

```python
train, test = train_test_split(df, test_size=0.2)
print(train)
print(test)
```

```python
y = train.label
x = [train.mean_R, train.mean_G, train.mean_B]
x = train.drop("label", axis=1)

print(x.shape)
print(y.shape)
```

```python
classifier = svm.SVC(kernel='poly')
classifier.fit(x,y)
```

```python
ytest = test.label
xtest = [test.mean_R, test.mean_G, test.mean_B]
xtest = test.drop('label', axis = 1)

ypred = classifier.predict(xtest)

acc = metrics.accuracy_score(ytest, ypred)
print(acc)
```
